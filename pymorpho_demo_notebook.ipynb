# Demonstrating pyMarshMorpho2D

This notebook demonstrates a Landlab component based on Giulio Mariotti's MarshMorpho2D.



Do some imports:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from landlab.io import read_esri_ascii
from landlab import imshow_grid
from pymarshmorpho2d import MarshEvolver
```

Create a Landlab grid by reading in a DEM, setting any NODATA nodes to be closed boundaries, so that their values are not considered in the plot range:

```python
(grid, topo) = read_esri_ascii('examples/zSW3.asc',
                               name='topographic__elevation')
grid.set_nodata_nodes_to_closed(topo, 999)
imshow_grid(grid, topo)
```

```python
mev = MarshEvolver(grid)
```
